# AFTER-TAX EARNINGS

Forward, after-tax earnings estimates in progressive taxation systems. 

After changing any parameters as you see fit, rerun all code cells to generate estimates.

#### LIBRARY IMPORTS

```python
import income_analyzer
from math import inf # upper bound to define bracket in the progressive taxation system
```

## PREPARING VARIABLES

### TAX BRACKET AND RATES

Below are the tax brackets and tax rates in Australia's progressive taxation system. Change these as you see fit.

```python
tax_brackets_AUD = [
    tuple([0, 18200]), 
    tuple([18200, 45000]), 
    tuple([45000, 120000]), 
    tuple([120000, 180000]), 
    tuple([180000, inf])
]
tax_rates_AUD = [
    0, 
    0.19, 
    0.325, 
    0.37, 
    0.45
]
```

### EARNINGS CLASS VARIABLES

```python
wage = 80000 # yearly wage
wage_growth = 8 # annualised percentage growth in yearly wage
savings_rate = 50 # percentage of after tax earnings saved
savings_returns = 7 # annualised returns earned on reinvesting savings
```

## EARNINGS CLASS

Pass variables to define the class `Earnings`.

```python
earnings = income_analyzer.Earnings(
    tax_brackets_AUD,
    tax_rates_AUD,
    wage,
    wage_growth,
    savings_rate,
    savings_returns
)
```

### METHODS

#### NET EARNINGS AFTER SOME NUMBER OF YEARS

```python
earnings.earningsAfter(10) # pass the number of years to generate estimate of after-tax earnings
```

#### YEARS TAKEN TO EARN TARGET

```python
tgt_earnings = 1000000 # pass future earnings to target to calculate number of years to earn target
earnings.yearsUntil(tgt_earnings)
```

#### EARNINGS PLOT FOR SOME NUMBER OF YEARS

```python
years = 20
earnings.plt(years) # pass years; plot earnings after year for year in years
```
